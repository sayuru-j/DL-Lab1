```python
import numpy as np
```

Creating a random 4x4 array from an exponential dist.

```python
random_array = np.random.exponential(scale=1.0, size=(4,4))
print(random_array)
```

Creating a random 100000x1 array from an exponential dist.

```python
random_array2 = np.random.exponential(scale=1.0, size=(100000,1))
print(random_array2)
```

Plot histograms

```python
import matplotlib.pyplot as plt

plt.hist(random_array2, density=True, bins=100, histtype="step", color="blue", label="Exponential")
# Uniform dist.
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="red", label="Uniform")
# Normal dist.
plt.hist(np.random.normal(size=100000), density=True, bins=100, histtype="step", color="green", label="Normal")
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

Plotting Z axis in a 3D plot

```python
x = np.linspace(-5, 5)
y = np.linspace(-5, 5)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X,Y,Z)
ax.set_xlabel('X', color="red")
ax.set_ylabel('Y', color="green")
ax.set_zlabel('Z', color="blue")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
df.head()
```

Scatter plot

```python
# Scatterplot arguments
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=True, # No regression line
           hue='Stage')   # Color by evolution stage
```

Box plot

```python
# Boxplot
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=df)
```

**pearson standard coefficient**

```python
features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
feature_data = df[features]

plt.figure(figsize=(9,8))
sns.heatmap(feature_data.corr(method="pearson"))
plt.title("Pearson Standard Coefficient")
```

**spearman rank correlation**

```python
sns.heatmap(feature_data.corr(method="spearman"))
plt.title("Spearman Rank Correalation")
```
